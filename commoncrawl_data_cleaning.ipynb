```
cc_data_dir = '/content/drive/MyDrive/cc-multilang-text-corpuses/hy/hy-cc'
sorted_batch_files = subprocess.getoutput('ls -S {}'.format(cc_data_dir)).split('\n')
files = ['{}/{}'.format(cc_data_dir, batch_file_name) for batch_file_name in sorted_batch_files]
```

```
start_from = 1100
go_to = 1299

for ten_batch in tqdm(range(start_from, go_to, 10)):
    command_string = 'timeout 60m python clear_a_batch.py --lang={lang} --batch_path={batch_path}'
    command = ' & '.join([command_string.format(batch_path = files[batch_id], lang = 'hy' ) for batch_id in range(ten_batch, ten_batch + 10)])
    
    !{command}
        
    !tar -czvf /content/hy-clean-{str(ten_batch)}-{str(ten_batch + 9)}.tgz ./clear/*.txt
    !rm /content/clear/*
    !cp /content/hy-clean-{str(ten_batch)}-{str(ten_batch + 9)}.tgz !ls /content/drive/MyDrive/cc-multilang-text-corpuses/hy/hy-clean/
```
